## Separating Categorical and Numerical Variables

```python
# Seprating out the categorical and continuous variables
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                # looping on the number of columns
        if df[column].dtype== object:     
                                                   
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         # Calling the function


table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

```python
import json
# Convert the string representation of production companies into a list of dictionaries
df['production_companies'] = df['production_companies'].apply(lambda x: json.loads(x))

```

```python
df.columns
```

```python
# # Initialize an empty dictionary to store production company names and their counts
production_counts = {}

# Iterate over each row in the DataFrame
for row in df.itertuples(index=False):
    # Iterate over each production company in the row
    for company in row.production_companies:
        # Extract the production company name
        company_name = company['name']
        
        # Increment the count for the production company
        if company_name in production_counts:
            production_counts[company_name] += 1
        else:
            production_counts[company_name] = 1

# Create a new DataFrame to store the production company names and their counts
production_df = pd.DataFrame({'Production Company': list(production_counts.keys()), 'Count': list(production_counts.values())})

# Sort the DataFrame by count in descending order
production_df = production_df.sort_values('Count', ascending=False)
production_df
```

```python
import statistics as st
```

```python

# Calculate the measures of central tendency using describe() function

ct=top_500_rows.loc[:,['budget','revenue','runtime']]
```

```python
Mean=[]
Median=[]
Mode=[]

for col in ct.head():
    Median.append(ct[col].median())  #calculating median value and store in Median list
    Mean.append(ct[col].mean())
    Mode.append(st.mode(ct[col]))

col_names=['Budget','Revenue','Runtime']
row_head=['Mean','Median','Mode']
 #Calculate a dataframe with all the values
    
d_data=pd.DataFrame([Mean,Median,Mode],columns=col_names)


d_data.insert(0,'Measures',row_head)
d_data
```
